```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
with open("/content/drive/My Drive/conversational_ai/JaneEyre.txt", "r") as file:
  text = file.read()
```

```python
wordcloud = WordCloud(
width=800,
height=400,
background_color="white",
stopwords=set(STOPWORDS),
min_font_size=10,
).generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
stopwords=set(STOPWORDS)
stopwords.update(["said","now"])

custom_wordcloud = WordCloud(
width=400,
height=800,
background_color="black",
stopwords=stopwords,
min_font_size=20,
max_words=10,
colormap="plasma"
).generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(custom_wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
mask_image = Image.open("/content/drive/My Drive/conversational_ai/batman_logo.png")
mask_image = mask_image.convert("RGBA")  # Convert to RGBA to handle transparency
data = mask_image.getdata()

# Change transparent areas to white (255, 255, 255, 255)
newData = []
for item in data:
    if item[3] == 0:  # Check if pixel is transparent
        newData.append((255, 255, 255, 255))  # Replace with white
    else:
        newData.append(item)

mask_image.putdata(newData)
mask = np.array(mask_image)
```

```python
mask_wordcloud = WordCloud(
width=900,
height=512,
background_color="white",
stopwords=stopwords,
min_font_size=5,
max_words=100,
colormap="inferno",
mask=mask
).generate(text)
```

```python
plt.figure(figsize=(10, 10), facecolor=None)  # Adjust figure size if needed
plt.imshow(mask_wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
